Nhập dữ liệu từ file excel:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
nk7=pd.read_excel('NK-T7.xlsx')
nk7['month']=7
nk9=pd.read_excel('NK-T9.xlsx')
nk9['month']=9
xk7=pd.read_excel('XK-T7.xlsx')
xk7['month']=7
xk9=pd.read_excel('XK-T9.xlsx')
xk9['month']=9
nk=pd.concat([nk7,nk9],ignore_index=True)
xk=pd.concat([xk7,xk9],ignore_index=True)
```

Tổng sản lượng nhập khẩu tháng 9 giảm mạnh.

```python
month_weight=sns.barplot(x='month',y='Tổng trọng lượng hàng (Gross)',data=nk,ci=None,estimator=sum)
for patch in month_weight.patches:
    month_weight.annotate(round(patch.get_height(),2),(patch.get_x()+patch.get_width()/2,patch.get_height()),ha='center',va='center',xytext=(0,5),textcoords = 'offset points')
month_weight.set_ylabel('Tổng trọng lượng gross (KGM)')
plt.title('Biểu đồ thể hiện tổng sản lượng nhập khẩu theo tháng')
plt.yscale('log')
```

Nhận thấy đồng tiền được sử dụng trong giao dịch nhập khẩu hàng hóa có sự đa dạng kể từ tháng 7 tới tháng 9.
Cụ thể là: Sử dụng thêm đồng THB, RUB, NOK, SGD
Mặt khác, lượng giao dịch đồng GBP có sự giảm mạnh (99.98%), thậm chí đồng AUD còn không được sử dụng trong giao dịch ở tháng 9.

```python
nk=nk.assign(total_import_value= lambda x : x['Trị giá tính thuế(S)']+x['Số tiền thuế nhập khẩu'])
# Visualize month vs total_import_value
month_value=sns.barplot(hue='month',y='total_import_value',data=nk,ci=None,x='Mã đồng tiền của đơn giá',estimator=sum)
month_value.set(ylabel='Tổng giá trị nhập khẩu',xlabel='Mã đồng tiền')
plt.title('Biểu đồ thể hiện tổng giá trị hàng nhập khẩu theo tháng và đơn vị tiền tệ')
plt.yscale('log')
```

Đồng EUR có lưu lượng giao dịch tăng mạnh 1130.61%, tiếp theo là các đồng HKD (201.95%) và CNY (61.59%)

```python
month_vs_total_value=nk.groupby(['month','Mã đồng tiền của đơn giá'],as_index=False).total_import_value.sum()
percent_check=month_vs_total_value[month_vs_total_value.month==7].merge(month_vs_total_value[month_vs_total_value.month==9],on='Mã đồng tiền của đơn giá',suffixes=['_7','_9'],how='inner').assign(percent_change=lambda x:(x['total_import_value_9']-x['total_import_value_7'])*100/x['total_import_value_7'])
percent_check
percent_check_plot=sns.barplot(x='Mã đồng tiền của đơn giá',y='percent_change',data=percent_check,estimator=sum)
for patch in percent_check_plot.patches:
    if patch.get_height() >0:
        percent_check_plot.annotate(str(round(patch.get_height(),2))+'%',(patch.get_x()+patch.get_width()/2,patch.get_height()),ha='center',va='center',xytext=(0,5),textcoords = 'offset points')
    else:
        percent_check_plot.annotate(str(round(patch.get_height(),2))+'%',(patch.get_x()+patch.get_width()/2,patch.get_height()),ha='center',va='center',xytext=(0,-5),textcoords = 'offset points')
plt.title('Biểu đồ thể hiện tỷ lệ giá trị nhập khẩu theo đơn vị tiền tệ so sánh giữa tháng 9 và tháng 7')
plt.xlabel('Mã đồng tiền')
```

Mặt hàng có mã HS 20079920 được nhập khẩu trong tháng 9 tăng mạnh so với tháng 7.
Mặt hàng có mã HS 20071000 được nhập khẩu trong tháng 9 giảm mạnh so với tháng 7.

```python
# Mô tả sản lượng nhập khẩu của từng mã HS [20071000, 20079990, 20079920, 20079100, 20079910]
data_hs=nk.groupby(['month','Mã số hàng hóa'],as_index=False).agg({'Tổng trọng lượng hàng (Gross)':'sum'})

percent_check=data_hs[data_hs.month==7].merge(data_hs[data_hs.month==9],on='Mã số hàng hóa',suffixes=['_7','_9'],how='inner').assign(percent_change=lambda x:(x['Tổng trọng lượng hàng (Gross)_9']-x['Tổng trọng lượng hàng (Gross)_7'])*100/x['Tổng trọng lượng hàng (Gross)_7'])

hs_fig=sns.barplot(y='percent_change',x='Mã số hàng hóa',data=percent_check,ci=None)
for patch in hs_fig.patches:
    if patch.get_height() >0:
        hs_fig.annotate(str(round(patch.get_height(),2))+'%',(patch.get_x()+patch.get_width()/2,patch.get_height()),ha='center',va='center',xytext=(0,5),textcoords = 'offset points')
    else:
        hs_fig.annotate(str(round(patch.get_height(),2))+'%',(patch.get_x()+patch.get_width()/2,patch.get_height()),ha='center',va='center',xytext=(0,-5),textcoords = 'offset points')
plt.title('Biểu đồ thể hiện tỷ lệ sản lượng nhập khẩu theo mã HS hàng hóa so sánh giữa tháng 9 và tháng 7')
```

# Phân tích xu hướng xuất nhập khẩu

Có thể thấy trong tháng 7 ta chỉ thu về được đồng VND.
Tỷ lệ và số lượng ngoại tệ thu về cũng tăng theo chiều tăng của tháng.
=> Xu hướng đa dạng hóa xuất nhập khẩu. Đẩy mạnh xuất khẩu hơn trong tương lai.

```python
#Cash Flow
nk_gộp=nk.groupby(['month','Mã đồng tiền của đơn giá'],as_index=False).total_import_value.sum()
xk_gộp=xk.groupby(['month','Mã đồng tiền của đơn giá hóa đơn'],as_index=False)['Trị giá hóa đơn'].sum()
xk_gộp.columns=['month','Mã đồng tiền của đơn giá','total_export_value']
cash_flow=xk_gộp.merge(nk_gộp,on=['month','Mã đồng tiền của đơn giá'],how='outer')
cash_flow.total_export_value=cash_flow.total_export_value.fillna(0)
cash_flow=cash_flow.assign(cash_flow=lambda x: (x.total_export_value-x.total_import_value)*100/x.total_import_value)
cash_flow.cash_flow=cash_flow.cash_flow.fillna(100.0)

sns.barplot(x='Mã đồng tiền của đơn giá',y='cash_flow',data=cash_flow,hue='month')
plt.title('Biểu đồ thể hiện tỷ lệ dòng tiền lãi thu về so với dòng tiền chi ra')
```
